```python
#%% Import relevant modules
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read in data
myData = pd.read_csv(
    "https://raw.githubusercontent.com/bachmannpatrick/Python-Class/master/data/transactions.csv")
```

```python
eu_date_format = "%d.%m.%Y"  # TransDate and Birthdate, e.g., 04.05.1987

#fix the Date format
myData["TransDate"] = pd.to_datetime(myData["TransDate"], format=eu_date_format)
```
