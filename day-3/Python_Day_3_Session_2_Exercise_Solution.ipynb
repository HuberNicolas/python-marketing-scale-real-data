###30 - Aggregate operations in SQL

0. Connect to the database again

```
sqlite_file="database.sqlite" # Path of SQLite-Database
con=sqlite3.connect(sqlite_file) # Build a connection to the sqlite database
# Get the cursor object for the connection to traverse the relational database.
cursor=con.cursor()
```

1.	Show for each customer the latest transaction date (LatestPurch) as well as the maximum purchase amount per transaction (MaxPurch). Please use  MAX(VariableName) here.

```
pd.read_sql_query("SELECT Customer, MAX(date(TransDate)) AS LatestPurch, MAX(PurchAmount) AS MaxPurch FROM transactions GROUP BY Customer;", con)
```

2.	Calculate the average purchase amount (AvgPurch) and average cost (AvgCost) per year. Please use AVG(VariableName) here.

```
#cursor.execute("PRAGMA table_info(transactions);") # Get column informations
#cursor.fetchall()
pd.read_sql_query("SELECT date(TransDate, 'start of year') AS Year, AVG(PurchAmount) AS AvgPurch, AVG(Cost) AS AvgCost FROM transactions GROUP BY Year;", con)
```

```
#Alternative: One can also refer to YEAR instead of transforming TransDate again
pd.read_sql_query("SELECT date(TransDate, 'start of year') AS Year, AVG(PurchAmount) AS AvgPurch, AVG(Cost) AS AvgCost FROM transactions GROUP BY Year;", con)
```

3.	Write a query for a table that shows the number of transactions (Transactions) and the average purchase amount (AvgPurch) per customer per month.

```
pd.read_sql_query("SELECT Customer, date(TransDate, 'start of month') AS MONTH, count(*) AS Transactions, AVG(PurchAmount) AS AvgPurch FROM transactions GROUP BY Customer, MONTH;", con)
```

```
con.close() #We are done. Let's close the database connection
```
