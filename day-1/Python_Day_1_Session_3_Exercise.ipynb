###Preparation

1.   Read the csv file and store it in the variable "myData". Note: this file is hostes in a GitHub repository. https://raw.githubusercontent.com/bachmannpatrick/Python-Class/master/data/transactions.csv

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
myData = pd.read_csv(filepath_or_buffer="https://raw.githubusercontent.com/bachmannpatrick/Python-Class/master/data/transactions.csv", sep=",")
```

```python
myData.head()
```

###10 - Creating basic plots

1. Create a boxplot for the variable PurchAmount(x).

```python
plt.boxplot(x=myData["PurchAmount"])
plt.show()
```

2. Create a scatter plot for the variables `Quantity`(x) and `PurchAmount`(y). Can you observe any correlation?

```python
plt.scatter(x=myData["Quantity"], y=myData["PurchAmount"])
plt.show()
```

###11 - Format basic plots

1.Format your earlier scatter plots in an appealing way.  Choose appropriate limits for the axes.  Consider changing colors, sizes, and fonts for data points and text labels.


```python
plt.plot(
    myData["Quantity"],
    myData["PurchAmount"],
    ".",
    label="Transaction",
   
)
plt.xlabel("Quantity", fontweight ='bold', color="green")
plt.ylabel("PurchAmount", fontweight ='bold', color="blue")
plt.title("PurchAmount by Quantity", style='italic', color="red")
#plt.xlim(0, myData["Quantity"].max())
plt.xlim(0, 20)
plt.ylim(0, myData["PurchAmount"].max())
plt.legend(loc="upper right", fontsize =12, title="Legend")

plt.show()
```

```python
import numpy as np
import seaborn as sns


# calculate correlation matrix
myData = myData.drop(["TransDate"], axis=1)
corr_matrix = myData.corr()

# plot heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
```

###12 - Adding further features to plots

1. Plot (a) the scatter plot  of Cost and `Purchase Amount` and (b) a historgram of `Purchase Amount` and in the same graph.



```python
plt.subplot(1, 2, 1)
plt.plot(
    myData["Cost"],
    myData["PurchAmount"],
    ".",
    label="Transaction",
   
)
plt.xlabel("Cost")
plt.ylabel("PurchAmount")
plt.title("PurchAmount by Cost")
plt.xlim(0, 20)
plt.ylim(0, 200)
plt.legend(loc="upper right", fontsize =12, title="Legend")


plt.subplot(1, 2, 2)
plt.hist(myData["PurchAmount"], color="green")
plt.title("Hist PurchAmount")
plt.xlabel("Purchase Amount")
plt.ylabel("Frequencey")
plt.show()
```
