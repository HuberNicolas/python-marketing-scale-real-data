```python
#load the Pandas package
import pandas as pd
#Read the csv file and store it in the variable "myData". Note: this file is hostes in a GitHub repository.
myData=pd.read_csv(filepath_or_buffer="https://raw.githubusercontent.com/bachmannpatrick/Python-Class/master/data/transactions.csv", sep=",")
#load matplotlib and pyplot as plt
import matplotlib.pyplot as plt
#load the package numpy as np
import numpy as np
#load the package seaborn as sbn
import seaborn as sns
```

###13 - Creating advanced plots

####**Slide:** Example 1: Create a histogram

```python
#load the package seaborn as sbn
import seaborn as sns
sns.distplot(myData["PurchAmount"], kde=False)

plt.show()
```

####**Slide:** Example 1: Flip the coordinates of the histogram

```python
sns.distplot(myData["PurchAmount"], kde=False, vertical=True)

plt.show()
```

####**Slide:** Example 2: Create a scatterplot

```python
sns.lmplot(x="PurchAmount", y="Cost", data=myData, fit_reg=False)
plt.show()
```

####**Slide:** Plot title

```python
#plot 1:
sns.distplot(myData["PurchAmount"], kde=False)
plt.title("Histogram of Purchase Amount")
plt.show()
```

```python
#plot2:
sns.lmplot(x="PurchAmount", y="Cost", data=myData, fit_reg=False)
plt.title("Costs by Purchase Amount")
plt.show()
```

####**Slide:** Axis lables

```python
#plot1
sns.distplot(myData["PurchAmount"], kde=False)
plt.title("Histogram of Purchase Amount")
plt.xlabel("Purchase Amount")
plt.show()
```

```python
#plot2:
sns.lmplot(x="PurchAmount", y="Cost", data=myData, fit_reg=False)
plt.title("Costs by Purchase Amount")
plt.xlabel("Purchase Amount")
plt.ylabel("Cost")
plt.show()
```

####**Slide:** Change point size

```python
sns.lmplot(x="PurchAmount", y="Cost", data=myData, fit_reg=False, scatter_kws={"s": 100})
plt.show()
```

###14 - Color palettes, themes, and styles


####**Slide:** Color do matter!

```python
#plot1
points=plt.scatter(myData["PurchAmount"], myData["Cost"], c=myData["Cost"], cmap="plasma")
plt.colorbar(points)
plt.xlabel("Purchase Amount")
plt.ylabel("Cost")
plt.show()
```

```python
#plot2
points=plt.scatter(myData["PurchAmount"], myData["Cost"], c=myData["Cost"], cmap="gray")
plt.colorbar(points)
plt.xlabel("Purchase Amount")
plt.ylabel("Cost")
plt.show()
```

####**Slide:** Using color palettes to indicate a third dimension

```python
sns.lmplot("PurchAmount", "Cost", data=myData, fit_reg=False,
hue="Quantity", palette="PuBuGn_d")
plt.plot()
```

###15 - Very advanced plotting topics

####**Slide:** Overlay plots on the same axes

```python
fig, ax = plt.subplots()
sns.regplot(x="PurchAmount", y="Cost", data=myData, fit_reg=False, ax=ax)
sns.regplot(x="PurchAmount", y="Cost", data=myData, order=2, scatter=False, ax=ax)
```

####**Slide:** With facets you can plot up to four dimensions in one single figure


```python
#in order to create a plot for every year, we need to extract the year in myData first and store in a new column.
myData["Year"] = pd.DatetimeIndex(myData["TransDate"]).year
```

```python
g = sns.FacetGrid(myData, col="Year", col_wrap=3)
g.map(plt.scatter, "PurchAmount", "Cost")
```

####**Slide:** Facets split up your data by one or more variables and plot the subsets of data together


```python
#in order to create a plot for every year, we need to extract the year in myData first and store in a new column.
myData["Year"] = pd.DatetimeIndex(myData["TransDate"]).year
```

```python
g = sns.FacetGrid(myData, col="Year", hue="Quantity", col_wrap=3)
g.map(plt.scatter, "PurchAmount", "Cost")
g.add_legend();
```
