df.(i)loc[ROW_INDEXER, COL_INDEXER]

iloc -> Integer based Selection => integers, either single, list or slice object (2, [2,5,6], slice object) / boolean arrays [False, True, False, False] as  input

loc -> Label based Selection => labels either single, list, or slice object (5 or 'a', ['a', 'cd'], slice object) / boolean arrays [False, True, False, False] as  input
-> Note that 5 is interpreted as a label of the index, and never as an integer position along the index.

The key difference between loc and iloc is in how they handle index and column labels versus integer positions:

Use loc when you want to select data based on labels(index or column names).
Use iloc when you want to select data based on integer positions(index or column positions)


###17 - Selection rows

1. Select rows 10 to 20 (inlcuding 20) from the `myData` DataFrame. *Hint:* Please have a careful look at the terminology. The row number of the first line is 1, the row index of the first line is 0.



```python
myData.iloc[0:21]
```

2. Select all purchases from 2010 from the myData DataFrame.




```python
myData.loc[lambda df: myData["TransDate"].dt.year == 2010]
```

```python
myData["TransDate"].dt.date
```

3. Select purchases made on 19.12.2010 or on 21.12.2010 from the `myData` DataFrame. *Hint:* Make sure the date variable is in the correct format.

```python
first_date = "19.12.2010"
second_date = "21.12.2012"


myData.loc[
    lambda df: 
           (myData["TransDate"].dt.date == pd.to_datetime(first_date, format="%d.%m.%Y").date()) 
           | 
           (myData["TransDate"].dt.date == pd.to_datetime(second_date, format="%d.%m.%Y").date())
           ]
```

4. Select all purchases with quantities between 2 and 5 (including 2 and 5) from the `myData` DataFrame.


```python
first_quantity = 2
second_quantity = 5


myData.loc[lambda df: 
           (myData["Quantity"] >= first_quantity) 
           & 
           (myData["Quantity"] <= second_quantity)
           ]
```

5. Select all purchases with purchase amount greater than 100 which were made from 01.01.2009 onwards from the `myData` DataFrame.


```python
deadline_date = "01.01.2009"
min_amount = 100


myData.loc[
    lambda df: 
           (myData["TransDate"].dt.date >= pd.to_datetime(deadline_date, format="%d.%m.%Y").date()) 
           & 
           (myData["PurchAmount"] > min_amount)
           ]
```
